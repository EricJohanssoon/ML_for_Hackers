Return a single element vector of just the email body.
This is a very simple approach, as we are only using words as features

```python
def getMsg(path):
    lines = open(path, encoding="latin-1").readlines()
    startIndex = lines.index('\n')
    return ''.join(lines[startIndex+1 : len(lines)])
```

```python
import nltk
nltk.download('words')
```

Create a TermDocumentMatrix (TDM) from the corpus of SPAM email.
The TDM control can be modified, and the sparsity level can be altered.  This TDM is used to create the feature set used to do train our classifier.

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

def getTDM(docs):
    vec = CountVectorizer(stop_words = 'english', token_pattern='[\S]+')
    bag_of_words = vec.fit_transform(docs)
    sum_words = bag_of_words.sum(axis=0) 
    #words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    #words_freq = sorted(words_freq, key = lambda x: x[1], reverse=True)
    #words_freq = words_freq[:50]
    
    df = pd.DataFrame(bag_of_words.toarray(), columns=vec.get_feature_names()) 
    return df
```

This function takes a file path to an email file and a string, the term parameter, and returns the count of that term in the email body.

```python
def countWord(part, term):
    msg = getMsg(part)
    count = msg.count(term)
    return count
```

```python
import os

dataSpamDir = 'data/spam/'
spamPaths = os.listdir(dataSpamDir)
spamPaths = [f'data/spam/{i}' for i in spamPaths]

allSpam = [getMsg(path) for path in spamPaths]
tdm = getTDM(allSpam)
tdm
```
