First code snippet


```python
import numpy as np
from plotnine import *
import pandas as pd
%matplotlib inline

x = np.arange(-10, 10, 0.01)
y = 1 - x*x + np.random.normal(size = len(x), scale = 5)

fig = (
    ggplot(data =  pd.DataFrame({'X': x, 'Y': y}), mapping=aes(x = 'X',  y = 'Y')) +
    geom_point() +
    geom_smooth(se = False)
)

fig
```

Second code snippet

```python
x_squared = x * x
```

Third code snippet

```python
fig = (
    ggplot(data =  pd.DataFrame({'X': x_squared, 'Y': y}), mapping=aes(x = 'X',  y = 'Y')) +
    geom_point() +
    geom_smooth(method = 'lm', se = False)
)

fig
```

Fourth code snippet

```python
import statsmodels.api as sm
from sklearn.metrics import r2_score

model = sm.OLS(y, x).fit()
predictions = model.predict(x) 

model.summary()
```
