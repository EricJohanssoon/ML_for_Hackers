Ninth code snippet

```python
performance = pd.DataFrame()

df = pd.DataFrame({'X': x, 'Y': y})
msk = np.random.rand(len(df)) < 0.8
train = df[msk]
test = df[~msk]

for degree in range(1, 15):
    coefficients = np.polyfit(train.X, train.Y, degree)
    p = np.poly1d(coefficients)
    
    predicted_values_train = p(train.X)
    rmse_train = np.sqrt(np.square(predicted_values_train - train.Y).mean())
    performance = performance.append({'Degree': degree, 'Data': 'Training', 'RMSE': rmse_train}, ignore_index=True)
    
    predicted_values_test = p(test.X)
    rmse_test = np.sqrt(np.square(predicted_values_test - test.Y).mean())
    performance = performance.append({'Degree': degree, 'Data': 'Test', 'RMSE': rmse_test}, ignore_index=True)
```

Tenth code snippet

```python
fig = (
    ggplot(data =  performance, mapping=aes(x = 'Degree',  y = 'RMSE', linetype = 'Data')) +
    geom_point() + 
    geom_line()
)

fig
```

Using Lasso Regression (L1)

```python
from sklearn import linear_model
from sklearn.cross_validation import train_test_split

performance = pd.DataFrame()
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=3)

for alpha in np.arange(0, 1, 0.05):
    clf = linear_model.Lasso(alpha=alpha)
    clf.fit(X_train.reshape(-1, 1), y_train)
    
    predicted_values_train = clf.predict(X_train.reshape(-1, 1))
    rmse_train = np.sqrt(np.square(predicted_values_train - y_train).mean())
    performance = performance.append({'Alpha': alpha, 'Data': 'Training', 'RMSE': rmse_train}, ignore_index=True)
    
    predicted_values_test = clf.predict(X_test.reshape(-1, 1))
    rmse_test = np.sqrt(np.square(predicted_values_test - y_test).mean())
    performance = performance.append({'Alpha': alpha, 'Data': 'Test', 'RMSE': rmse_test}, ignore_index=True)
    
fig = (
    ggplot(data =  performance, mapping=aes(x = 'Alpha',  y = 'RMSE', linetype = 'Data')) +
    geom_point() + 
    geom_line()
)

fig
```
